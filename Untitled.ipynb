```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Define the folder path
folder_path = os.path.expanduser("~/Downloads/vestibular_mismatch_video2_data")

# Get a list of all files in the folder
files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

# Loop through each file and plot only "center.x" and "center.y"
for file in files:
    file_path = os.path.join(folder_path, file)

    try:
        # Try reading as a CSV
        df = pd.read_csv(file_path)

        # Check if the required columns exist
        columns_to_plot = ["center.x", "center.y"]
        available_columns = [col for col in columns_to_plot if col in df.columns]

        if available_columns:
            plt.figure(figsize=(8, 4))
            for column in available_columns:
                plt.plot(df[column], label=column)

            plt.title(f"{file} - Center Coordinates")
            plt.xlabel("Index")
            plt.ylabel("Value")
            plt.legend()
            plt.grid(True)
            plt.show()
        else:
            print(f"{file} does not contain 'center.x' or 'center.y'.")

    except Exception as e:
        print(f"Could not process {file}: {e}")


```
