```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib widget
SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

# Define file paths
data_path1 = "/Volumes/sambashare/data/photometry/Cohort3_test/B6J2780_processedData/photometry/Processed_fluorescence.csv"
data_path2 = "/Volumes/sambashare/data/photometry/Cohort1_test/B6J2717_processedData/photometry/Processed_fluorescence.csv"
data_path3 = "/Volumes/sambashare/data/photometry/Cohort1_test/B6J2718_processedData/photometry/Processed_fluorescence.csv"
data_path4 = "/Volumes/sambashare/data/photometry/Cohort1_test/B6J2721_processedData/photometry/Processed_fluorescence.csv"
data_path5 = "/Volumes/sambashare/data/photometry/Cohort1_test/B6J2722_processedData/photometry/Processed_fluorescence.csv"
data_path6 = "/Volumes/sambashare/data/photometry/Cohort3_test/B6J2781_processedData/photometry/Processed_fluorescence.csv"
data_path7 = "/Volumes/sambashare/data/photometry/Cohort3_test/B6J2783_processedData/photometry/Processed_fluorescence.csv"

# Load data
df1 = pd.read_csv(data_path1, index_col="TimeStamp")
df2 = pd.read_csv(data_path2, index_col="TimeStamp")
df3 = pd.read_csv(data_path3, index_col="TimeStamp")
df4 = pd.read_csv(data_path4, index_col="TimeStamp")
df5 = pd.read_csv(data_path5, index_col="TimeStamp")
df6 = pd.read_csv(data_path6, index_col="TimeStamp")
df7 = pd.read_csv(data_path7, index_col="TimeStamp")

# Plot
plt.figure(figsize=(10, 5))
plt.plot(df1.index, df1["dfF_470"], label=data_path1, color='black')
plt.plot(df2.index, df2["dfF_470"], label=data_path2, color='red', alpha =0.5)
plt.plot(df3.index, df3["dfF_470"], label=data_path3, color='blue', alpha = 0.5)
# plt.plot(df4.index, df4["dfF_470"], label=data_path4, color='magenta', alpha = 0.5)
plt.plot(df6.index, df6["dfF_470"], label=data_path6, color='magenta')
plt.plot(df7.index, df7["dfF_470"], label=data_path7, color='green')

# Formatting
plt.xlabel("time (seconds")
plt.ylabel("dfF")
plt.title("Time Series Comparison")
plt.legend()
plt.grid(True)
plt.show()
```
